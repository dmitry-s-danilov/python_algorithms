```python
size_range = 0, 101

allocated = 0
for new_size in range(*size_range):
    if allocated < new_size:
        new_allocated = 1 + (new_size >> 3) + (3 if new_size < 9 else 6)
    else:
        new_allocated = 0
    allocated += new_allocated
    print(f'{new_size} {new_allocated} {allocated}')
```
