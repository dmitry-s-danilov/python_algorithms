```python
from sys import getsizeof
```

```python
r = [(ord('A'), ord('Z') + 1),
     (ord('a'), ord('z') + 1),
     (ord('А'), ord('Я') + 1),
     (ord('а'), ord('я') + 1)]

u = [range(*_) for _ in r]

v = list()
for _ in u:
    v.extend(list(_))

w = {_: chr(_) for _ in v}

for i, c in w.items():
    print(f'{i:04}: {c} {getsizeof(c)}')
```
